```python
from pathlib import Path
import numpy as np
from scipy.io import loadmat
import xarray as xr
from datetime import datetime, timedelta
import seaborn as sns
import cmocean as co
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import colors as c
import seawater as sw
import xorca_brokenline as bl
```

```python
## Parameters
# Project path
project_path = Path.cwd() / '..' / '..' 
project_path = project_path.resolve()

#section lonlat file
sectionPath = Path('data/processed/sections/')
```

```python
sectionTransportFilename=[
    'osnapSection2000to09.nc',
    'osnapSection2010to18.nc'
]

for i,filename in enumerate(sectionTransportFilename):
    sectiont=xr.open_dataset(project_path / sectionPath / filename)
    if i == 0:
        section=sectiont
    else:
        section=xr.concat([section,sectiont],dim='time_counter')
        
```

```python
transport_37wto0_z200 = bl.calculate_transport(section.isel(c=slice(645,None,1)).isel(z=slice(0,14,1)))
transport_30wto0_z200 = bl.calculate_transport(section.isel(c=slice(805,None,1)).isel(z=slice(0,14,1)))
transport_37wto19w_z200 = bl.calculate_transport(section.isel(c=slice(645,1024,1)).isel(z=slice(0,14,1)))
transport_30wto19w_z200 = bl.calculate_transport(section.isel(c=slice(805,1024,1)).isel(z=slice(0,14,1)))
transport_37wto0_z500 = bl.calculate_transport(section.isel(c=slice(645,None,1)).isel(z=slice(0,19,1)))
transport_30wto0_z500 = bl.calculate_transport(section.isel(c=slice(805,None,1)).isel(z=slice(0,19,1)))
transport_37wto19w_z500 = bl.calculate_transport(section.isel(c=slice(645,1024,1)).isel(z=slice(0,19,1)))
transport_30wto19w_z500 = bl.calculate_transport(section.isel(c=slice(805,1024,1)).isel(z=slice(0,19,1)))
transport_37wto0_z800 = bl.calculate_transport(section.isel(c=slice(645,None,1)).isel(z=slice(0,22,1)))
transport_30wto0_z800 = bl.calculate_transport(section.isel(c=slice(805,None,1)).isel(z=slice(0,22,1)))
transport_37wto19w_z800 = bl.calculate_transport(section.isel(c=slice(645,1024,1)).isel(z=slice(0,22,1)))
transport_30wto19w_z800 = bl.calculate_transport(section.isel(c=slice(805,1024,1)).isel(z=slice(0,22,1)))
transport_37wto0_z1000 = bl.calculate_transport(section.isel(c=slice(645,None,1)).isel(z=slice(0,24,1)))
transport_30wto0_z1000 = bl.calculate_transport(section.isel(c=slice(805,None,1)).isel(z=slice(0,24,1)))
transport_37wto19w_z1000 = bl.calculate_transport(section.isel(c=slice(645,1024,1)).isel(z=slice(0,24,1)))
transport_30wto19w_z1000 = bl.calculate_transport(section.isel(c=slice(805,1024,1)).isel(z=slice(0,24,1)))
```

```python
def plot_section():    
    p = plt.figure(figsize=(15,5))
    # whole section
#     section['u_normal'].isel(time_counter=102).plot(cmap=co.cm.balance)
#     section['mask'].where(section['mask']==0).mean('time_counter').plot(cmap=c.ListedColormap(['gray']),add_colorbar=0)
    # slice parts of section slice(start,stop,step)
    section.isel(c=slice(785,-1,1))['u_normal'].mean('time_counter').plot()
    section.isel(c=slice(785,-1,1))['mask'].where(section['mask']==0).mean('time_counter').plot(cmap=c.ListedColormap(['gray']),add_colorbar=0)
    plt.ylim(4000,0)
    plt.ylabel("Depth (m)")
    plt.xlabel("Distance (m)")
    plt.title("4-year, 2014-2017, mean velocity normal to section. Positive into page.")
    return
```
